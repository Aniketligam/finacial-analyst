Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

load the csv file

```python

data = pd.read_csv(r"C:\Users\ANIKET LIGAM\Downloads\Financial Analytics data (1).csv")

# Display the first few rows of the dataset
print(data.head())
```

data cleaning 

```python
data_clean=data.drop(columns=["Unnamed: 4"],inplace=True)
data_clean=data.dropna(subset=["Name","Mar Cap - Crore",  "Sales Qtr - Crore"])

data_clean.info()
data_clean.head()
```

Descriptive Statistics

```python
Descriptive_stats=data_clean.describe()
print(Descriptive_stats)
```

Visualization

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10,6))
sns.scatterplot(x="Mar Cap - Crore",y="Sales Qtr - Crore",data=data_clean)
plt.title("Mar Cap vs Qtr sales")
plt.xlabel("Mar Capital")
plt.ylabel("Qtr Sales")
plt.show()
```

Identify Top Companies

```python
# Top 10 companies by Market Capitalization

Top_market_capital=data_clean.nlargest(10,"Mar Cap - Crore")
print(f"Top 10 Companies by Market Capitalization:")
print(Top_market_capital)
```

```python
# Top 10 companies by Quarterly Sales

Top_Quaterly_Sales=data_clean.nlargest(10,"Sales Qtr - Crore")
print(f"Top 10 Companies by Quarterly Sales:")
print(Top_Quaterly_Sales)
```
